```python
import numpy as np
```

```python
# Input matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8, 9],
              [10, 11, 12]])
```

```python
# Addition
print("A + B:\n", A + B)
```

```python
# Subtraction
print("A - B:\n", A - B)
```

```python
# Element-wise Multiplication
print("A * B (Element-wise):\n", A * B)
```

```python
print(A.shape)
print(B.shape)
```

```python
# Matrix Multiplication (if dimensions allow)

if A.shape[1] == B.shape[0]:
    print("A x B (Matrix Multiplication):\n", np.dot(A, B))
else:
    print("Matrix multiplication not possible.")
```

```python
# Transpose
print("Transpose of A:\n", A.T)
print("\nTranspose of B:\n", B.T)
```
